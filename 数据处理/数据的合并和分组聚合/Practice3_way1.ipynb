```python
#统计出2015到2017年25万条911的紧急电话的数据中不同类型的紧急情况的次数
df = pd.read_csv("./911.csv")
```

```python
#获取灾害类型
temp_list = df["title"].str.split(":").tolist()
type_list = list(set([i[0] for i in temp_list]))
```

```python
#构造全为0的数组
zero_df = pd.DataFrame(np.zeros((df.shape[0],len(type_list))),columns=type_list)
```

```python
#将对应位置填充为1
#方法一
for type in type_list:
    zero_df[type][df["title"].str.contains(type)] = 1

#方法二(笨拙)
# for i in range(df.shape[0]):
#     zero_df.loc[i,temp_list[i][0]] = 1
```
