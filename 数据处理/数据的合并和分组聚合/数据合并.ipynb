```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.ones((2,4)),index=["A","B"],columns=list("abcd"))

df2 = pd.DataFrame(np.zeros((3,3)),index=["A","B","C"],columns=list("xyz"))
```

```python
df1
```

```python
df2
```

```python
#以行索引合并，以df1为主
df1.join(df2)
```

```python
#以行索引合并，以df2为主
df2.join(df1)
```

```python
df3 = pd.DataFrame(np.ones((3,4)),index=["A","B","C"],columns=["M","N","O","P"])
df3.loc[:,"O"] = ["a","b","c"]

df4 = pd.DataFrame(np.ones((2,5)),index=["A","B"],columns=list("VWXYZ"))
df4.loc[:,"X"] = ["c","d"]
```

```python
df3
```

```python
df4
```

```python
#取交集(默认为inner)
df3.merge(df4,left_on="O",right_on="X",how="inner")
```

```python
#外连接
df3.merge(df4,left_on="O",right_on="X",how="outer")
```

```python
#左连接
df3.merge(df4,left_on="O",right_on="X",how="left")
```

```python
#右连接
df3.merge(df4,left_on="O",right_on="X",how="right")
```
