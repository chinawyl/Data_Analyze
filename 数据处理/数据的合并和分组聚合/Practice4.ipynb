```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
df = pd.read_csv("./911.csv")
```

```python
df
```

```python
#不同月份不同类型紧急电话的次数的变化情况
df["timeStamp"] = pd.to_datetime(df["timeStamp"])
df.set_index("timeStamp",inplace=True)
```

```python
df
```

```python
count_month = df.resample("M").count()["title"]
```

```python
count_month
```

```python
#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#x,y
_x = count_month.index
_y = count_month.values

#精确时间
_x = [i.strftime("%Y%m%d") for i in _x]

#绘制图片
plt.plot(range(len(_x)),_y)

#设置x轴的刻度
plt.xticks(range(len(_x)),_x,rotation=45)

#绘制表格
plt.grid(alpha=1) #alpha是表格透明度

#添加描述信息
plt.xlabel("时间",fontproperties=my_font)
plt.ylabel("紧急电话的次数",fontproperties=my_font)
plt.title("不同月份不同类型紧急电话的次数的变化情况",fontproperties=my_font)

#展示图片
plt.show()
```
