```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_excel('../实验数据/某药店销售数据.xlsx')
```

```python
data
```

```python
# 数据预处理

data.dropna(how='any',inplace=True)
data = data[data['应收金额']>data['实收金额']]
data = data[data['销售数量']>0]
data['总消费金额'] = data['销售数量']*data['实收金额']
data['消费次数'] = data['实收金额']/data['实收金额']
data['购药年月'] = [i[:7] for i in data['购药时间']]
data = data.loc[:,['购药时间', '购药年月', '社保卡号', '商品编码', '商品名称', '销售数量', '消费次数', '应收金额', '实收金额', '总消费金额']]
list_index = [i for i in range(len(data))]
data.index = list_index
```

```python
data
```

```python
# 按照年月分组

grouped = data.groupby(by='购药年月')
datas = grouped.sum()
datas
```

```python
#月均消费次数
datas['消费次数'].sum()/len(datas)
```

```python
#月均消费金额
datas['实收金额'].sum()/len(datas)
```

```python
#人均消费金额
datas['实收金额'].sum()/datas['消费次数'].sum()
```
