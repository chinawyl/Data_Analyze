```python
data = pd.read_excel('../实验数据/股票数据.xlsx',index_col='序号')
```

```python
data.head()
```

```python
data.X3 = data.X3.map(lambda x:x.strip('亿')).astype(float)
```

```python
#将X2，X3的万元和亿元单位删除并将X2的万与亿统一为亿
list_money = []
for i in data['X2']:
    if i[-2] == '万':
        list_money.append(float(i[:-2])/10000)
    elif i[-2] == '亿':
        list_money.append(float(i[:-2]))
data.X2 = list_money
data.X3 = data.X3.map(lambda x:x.strip('亿')).astype(float)
```

```python
data.head()
```

```python
#取出对应列并用库函数归一化，并查看相关性
data1 = data[['y','X1','X2','X3','X4']]

model = StandardScaler()
data2 = pd.DataFrame(model.fit_transform(data1))

data2.corr()
```

```python
# 2.等频离散化


```

```python
k = 4

w = [i/k for i in range(k+1)] #[0.0, 0.25, 0.5, 0.75, 1.0]

des = data.describe(percentiles=w)[4:4+k+1] #percentiles=w包含0%，原本只有25%，50%，75%，100%

pd.cut(data['肝气郁结证型系数'], bins = des.iloc[:,0].values, labels=range(k)) #bins要给一个numpy，不能直接给一个Series
```
