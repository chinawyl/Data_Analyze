```python
from matplotlib import pyplot as plt
```

```python
x = range(2,26,2)
```

```python
y = [15,13,14.5,17,20,25,26,26,27,22,18,15]
```

```python
#设置图片大小
plt.figure(figsize=(12,6),dpi=80)

#设置x轴的刻度
plt.xticks(x)

#绘制图片
plt.plot(x,y)

#保存图片
plt.savefig("./Test.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```

```python
#设置图片大小
plt.figure(figsize=(12,6),dpi=80)

#设置x轴的刻度
list_x = [i/2 for i in range(4,49)]
plt.xticks(list_x[::3])

#绘制图片
plt.plot(x,y)

#保存图片
plt.savefig("./Test1.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```

```python
#设置图片大小
plt.figure(figsize=(12,6),dpi=80)

#设置x轴的刻度
list_x = [i/2 for i in range(4,49)]
plt.xticks(list_x[::3])

#设置y轴的刻度
plt.yticks(range(min(y),max(y)+1))

#绘制图片
plt.plot(x,y)

#保存图片
plt.savefig("./Test2.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```
