```python
from matplotlib import pyplot as plt
```

```python
#设置图片大小
plt.figure(figsize=(16,12),dpi=80)

#用来正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']

labels = ['娱乐','育儿','饮食','房贷','交通','其它']

sizes = [2,5,12,70,2,9]
explode = (0,0,0,0.1,0,0)

plt.pie(sizes #每一块饼图的比例，为必填项
        ,explode=explode #每一块饼图 离开中心距离，默认值为（0,0），就是不离开中心
        ,labels=labels #每一块饼图外侧显示的说明文字
        ,autopct='%1.1f%%' #格式化
        ,shadow=False #阴影
        ,startangle=90 #起始绘制角度，默认图是从x轴正方向逆时针画起，如设定startangle=90则从y轴正方向画起
#         ,colors=['r','g','y','b','r','g'] #默认为None
       ) 

#添加图例
plt.legend(loc="upper right",fontsize=10,bbox_to_anchor=(1.1,1.05),borderaxespad=0.3)

#添加标签
plt.title("某月份家庭支出")

#保存图片
plt.savefig("./Test.png",dpi=200,bbox_inches='tight')

#展示图片
plt.show()
```
