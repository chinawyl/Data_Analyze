```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
file_path = "./IMDB-Movie-Data.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
#rating，runtime的分布情况

#准备数据
runtime_data = df["Runtime (Minutes)"].values
runtime_max = runtime_data.max()
runtime_min = runtime_data.min()

#组距
d = 5

#计算组数
num_bins = (runtime_max-runtime_min) // d

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#绘制图片
plt.hist(runtime_data,num_bins)

#设置x轴的刻度
plt.xticks(range(min(runtime_data),max(runtime_data)+d,d))

#添加描述信息
plt.xlabel("电影时长",fontproperties=my_font)
plt.ylabel("电影数",fontproperties=my_font)
plt.title("1000部电影时长频数直方图",fontproperties=my_font)

#绘制表格
plt.grid()

#展示图片
plt.show()
```
