```python
# 4.正态分布3σ原则


```

```python
data = pd.read_excel('../实验数据/led测试数据.xls',index_col=0)
```

```python
data.head()
```

```python
data.describe()
```

```python
signa3 = data.describe().loc['std'] * 3
mean = data.describe().loc['mean']
upper = mean + signa3
lower = mean - signa3
for name,value in data.iterrows():
    s = (value < lower) | (value > upper)
    if s.any(): # 如果当iterable所有的值都是0、''或False时，那么结果为False，如果所有元素中有一个值非0、''或False，那么结果就为True
        print('%s号产品不合格'%name)
```

```python
# 箱型图公式查找水质不合格地点和指标
data = pd.read_excel('../实验数据/水质检测数据.xlsx',index_col=1,header=0)
IQR = (data.describe().loc['75%']-data.describe().loc['25%'])
Qu = data.describe().loc['75%'] + 1.5*IQR
Ql = data.describe().loc['25%'] - 1.5*IQR
for name,value in data.iterrows():
    s = (value < Ql) | (value > Qu)
    if s.any():
        indexname = s[s==True].index[0]
        print('%s水质不合格,%s指标超标'%(name,indexname))
```
