```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_excel('../实验数据/catering_sale_all.xls',index_col=0,header=0)
```

```python
data.head()
```

```python
data.info()
```

```python
#相关性函数corr
data.corr() #默认为spearson(Pearson相关系数),可改为spearman(Spearman秩相关系数)
```

```python
#散点图矩阵
from pandas.plotting import scatter_matrix
scatter_matrix(data,figsize=(10,10))
plt.show()
```
