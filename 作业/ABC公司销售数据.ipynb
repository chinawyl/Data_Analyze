```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_excel('./ABC公司销售数据.xls')
```

```python
data
```

```python
grouped = data.groupby(by='省份代码')
data_new = grouped.sum()
data_new = data_new.sort_values(by='净销售额',ascending=False)
```

```python
data_new
```

```python
datas = data_new['净销售额']
p = datas.cumsum()/datas.sum()
key = p[p>0.8].index[0]  
key_num = datas.index.tolist().index(key)
datas
```

```python
plt.figure(figsize=(20,8),dpi=80)
datas.plot(kind ='bar')#绘制条形图
plt.ylabel('净销售额')
p.plot(color ='r',secondary_y = True,style = '-o',linewidth = 2)
plt.annotate(format(p[key_num],'.4%'), xy = (key_num,p[key_num]),xytext = (key_num*0.9,p[key_num]*0.9),
             arrowprops = dict(arrowstyle="->",connectionstyle = "arc3,rad=.2"))
#添加注释，即85%处的标记，这里包括了指定箭头的样式
plt.ylabel('净销售额占比')
plt.savefig("ABC公司销售数据.png")
plt.show()
```
