```python
from matplotlib import pyplot as plt
```

```python
#
plt.figure(figsize=(16,12),dpi=80)

#
plt.rcParams['font.sans-serif']=['SimHei']

labels = ['','','','','','']

sizes = [2,5,12,70,2,9]
explode = (0,0,0,0.1,0,0)

plt.pie(sizes #
        ,explode=explode # 0,0
        ,labels=labels #
        ,autopct='%1.1f%%' #
        ,shadow=False #
        ,startangle=90 #xstartangle=90y
#         ,colors=['r','g','y','b','r','g'] #None
       ) 

#
plt.legend(loc="upper right",fontsize=10,bbox_to_anchor=(1.1,1.05),borderaxespad=0.3)

#
plt.title("")

#
plt.savefig("./Test.png",dpi=200,bbox_inches='tight')

#
plt.show()
```
