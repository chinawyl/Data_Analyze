```python
from matplotlib import pyplot as plt
import random
from matplotlib.font_manager import FontProperties
```

```python
x = range(120)
```

```python
y = [random.randint(20,35) for i in range(120)]
```

```python
#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#设置x轴的刻度
x_labels = ["10点{}分".format(i) for i in range(60)]
x_labels += ["11点{}分".format(i) for i in range(60)]

#设置步长和旋转度数和中文
plt.xticks(list(x)[::3],x_lables[::3],rotation=45,fontproperties=my_font)

#添加描述信息
plt.xlabel("时间",fontproperties=my_font)
plt.ylabel("温度(摄氏度)",fontproperties=my_font)
plt.title("10点到12点每一分钟温度的变化情况",fontproperties=my_font)

#绘制图片
plt.plot(x,y)

#保存图片
plt.savefig("./Test_Practice.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```
