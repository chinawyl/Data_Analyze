```python
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
interval = [0,5,15,20,25,30,35,40,45,60,90]
width = [5,5,5,5,5,5,5,5,5,15,30,60]
quantity = [836,2737,3723,3926,3596,1438,3273,642,824,613,215,47]
```

```python
#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#绘制图片
plt.bar(range(12),quantity,width=1)

#设置x轴的刻度
_x=[i-0.5 for i in range(13)]
_x_labels = interval+[150]
plt.xticks(_x,_x_labels)

#绘制表格
plt.grid()

#添加描述信息
plt.xlabel("上班地点到家的时间",fontproperties=my_font)
plt.ylabel("人数",fontproperties=my_font)
plt.title("伪直方图",fontproperties=my_font)

#保存图片
plt.savefig("./Practice.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```
