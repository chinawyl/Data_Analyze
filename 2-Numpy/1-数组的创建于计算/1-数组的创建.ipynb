```python
import numpy as np
import random
```

```python
#默认数据类型为
a = np.array([1,2,3])
print(a)
print(type(a))
print()

b = np.array(range(10))
print(b)
print(type(b))
print()

c = np.arange(12)
print(c)
print(type(c))
print(c.dtype)
print()

#float数据类型
d = np.array(range(2,5),dtype="float32")
print(d)
print(d.dtype)
print()

#bool数据类型
e = np.array([1,1,0,0,1,0,1,0],dtype="bool")
print(e)
print(e.dtype)
print()

#调整数据类型
f = e.astype("int8")
print(f)
print(f.dtype)
print()

#小数位数
g = np.array([random.random() for i in range(10)])
print(g)
print(g.dtype)
print()

h = np.round(g,2)
print(h)
print()

#多于一个维度
i = np.array([[1,2],[3,4]])
print(i)
print()

#empty创建，numpy.empty(shape, dtype=float, order=‘C’)
j = np.empty([2,2,3],dtype=int)
print(j)
print()

#创建全为0或全为1的数组
print(np.zeros([3,4],dtype=float))
print()
print(np.ones([3,4],dtype=float))
print()

#创建等差数列
print(np.linspace(1,10,20)) #第一个参数为开始，第二个参数为结束，第三个参数为数量
```
