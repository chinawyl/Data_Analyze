```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
df = pd.read_csv("./BeijingPM20100101_20151231.csv")
```

```python
df
```

```python
period = pd.PeriodIndex(year=df["year"],month=df["month"],day=df["day"],hour=df["hour"],freq="H")
```

```python
period
```

```python
type(period)
```

```python
df["datetime"] = period
```

```python
df.set_index("datetime",inplace=True)
```

```python
df
```

```python
#进行降采样
df = df.resample("7D").mean()
```

```python
#删除缺失值
data = df["PM_US Post"].dropna()
data_china = df["PM_Dongsi"].dropna()
```

```python
#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#x,y
_x = data.index
_x_china = [i.strftime("%Y%m%d") for i in data_china.index]
_y = data.values
_y_china = data_china.values

#精确时间
_x = [i.strftime("%Y%m%d") for i in _x]

#绘制图片
plt.plot(range(len(_x)),_y,label="US_Post")
plt.plot(range(len(_x_china)),_y_china,label="CN_Post")

#设置x轴的刻度
plt.xticks(range(0,len(_x),10),list(_x)[::10],rotation=45)

#绘制表格
plt.grid(alpha=1) #alpha是表格透明度

#添加描述信息
plt.xlabel("时间",fontproperties=my_font)
plt.ylabel("PM2.5数值",fontproperties=my_font)
plt.title("中国和美国PM2.5数据对比",fontproperties=my_font)

#添加图例
plt.legend(loc="best")

#展示图片
plt.show()
```
