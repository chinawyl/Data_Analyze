```python
import numpy as np
import pandas as pd
```

```python
#统计函数


```

```python
#pct_change():将每个元素与其前一个元素进行比较，并计算变化百分比
#可用于股票涨幅
s = pd.Series([1,2,3,4,5,4])
print (s.pct_change())

df = pd.DataFrame(np.random.randn(5, 2))
print (df.pct_change())
```

```python
#协方差:cov
frame = pd.DataFrame(np.random.randn(10, 5), columns=['a', 'b', 'c', 'd', 'e'])
print (frame['a'].cov(frame['b']))
print (frame.cov())
```

```python
#相关性:corr
frame = pd.DataFrame(np.random.randn(10, 5), columns=['a', 'b', 'c', 'd', 'e'])

print (frame['a'].corr(frame['b']))
print (frame.corr())
```

```python
#数据排名:rank
'''
Rank支持不同的tie-breaking方法，用方法参数指定 -

average - 并列组平均排序等级min - 组中最低的排序等级max - 组中最高的排序等级first - 按照它们出现在数组中的顺序分配队列


'''
s = pd.Series(np.random.np.random.randn(5), index=list('abcde'))

s['d'] = s['b'] # so there's a tie

print (s.rank())
```

```python
#窗口函数


```

```python
#rolling()
#由于窗口大小为3(window)，前两个元素有空值，第三个元素的值将是n，n-1和n-2元素的平均值。这样也可以应用上面提到的各种函数了
#滚动求平均，第三列为第一、二、三列求平均，第四列为第二、三、四列求平均

df = pd.DataFrame(np.random.randn(10, 4),
index = pd.date_range('1/1/2020', periods=10),
columns = ['A', 'B', 'C', 'D'])

print (df.rolling(window=3).mean())
```

```python
#expanding()
df = pd.DataFrame(np.random.randn(10, 4),
      index = pd.date_range('1/1/2018', periods=10),
      columns = ['A', 'B', 'C', 'D'])
print (df.expanding(min_periods=3).mean())
```

```python
#ewm():加权求平均
#指定com，span，halflife参数，并在其上应用适当的统计函数。它以指数形式分配权重。
df = pd.DataFrame(np.random.randn(10, 4),
   index = pd.date_range('1/1/2019', periods=10),
   columns = ['A', 'B', 'C', 'D'])
print (df.ewm(com=0.5).mean())
```

```python
#聚合函数


```

```python
#在整个数据框上应用聚合
df = pd.DataFrame(np.random.randn(10, 4),
      index = pd.date_range('1/1/2019', periods=10),
      columns = ['A', 'B', 'C', 'D'])

print("=======================================")
print(df)
r = df.rolling(window=3,min_periods=1) #小于3行就用1行或两行计算平均
print (r)
print("==================使用aggregate=====================")
print(r.aggregate(np.mean))
print("==================使用agg=====================") #更自由
print(r.agg(np.mean))
print("==================rolling.mean=====================")
print(r.mean)
```

```python
#数据框的单个列上应用聚合
df = pd.DataFrame(np.random.randn(10, 4),
      index = pd.date_range('1/1/2000', periods=10),
      columns = ['A', 'B', 'C', 'D'])
print (df)
print("====================================")
r = df.rolling(window=3,min_periods=1)
print (r['A'].aggregate(np.sum))
```

```python
#在DataFrame的多列上应用聚合
df = pd.DataFrame(np.random.randn(10, 4),
      index = pd.date_range('1/1/2018', periods=10),
      columns = ['A', 'B', 'C', 'D'])
print (df)
print ("==========================================")
r = df.rolling(window=3,min_periods=1)
print (r[['A','B']].aggregate(np.sum))
```

```python
#在DataFrame的单个列上应用多个函数
df = pd.DataFrame(np.random.randn(10, 4),
      index = pd.date_range('2019/01/01', periods=10),
      columns = ['A', 'B', 'C', 'D'])
print (df)

print("==========================================")

r = df.rolling(window=3,min_periods=1)
print (r['A'].aggregate([np.sum,np.mean]))
```

```python
#在DataFrame的多列上应用多个函数
df = pd.DataFrame(np.random.randn(10, 4),
      index = pd.date_range('2020/01/01', periods=10),
      columns = ['A', 'B', 'C', 'D'])

print (df)
print("==========================================")
r = df.rolling(window=3,min_periods=1)
print (r[['A','B']].aggregate([np.sum,np.mean]))
```

```python
#将不同的函数应用于DataFrame的不同列
df = pd.DataFrame(np.random.randn(3, 4),
      index = pd.date_range('2020/01/01', periods=3),
      columns = ['A', 'B', 'C', 'D'])
print (df)
print("==========================================")
r = df.rolling(window=3,min_periods=1)
print (r.aggregate({'A' : np.sum,'B' : np.mean}))
```
