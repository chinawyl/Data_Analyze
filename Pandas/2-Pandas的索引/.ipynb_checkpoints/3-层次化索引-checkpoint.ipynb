```python
import numpy as np
import pandas as pd
import random
```

```python
#隐式构造


```

```python
data = np.random.randint(0,100,size=4)
s = pd.Series(data,index=[['期中','期中','期末','期末'],['语文','数学','语文','数学']])
print(s)
print()
print(s.loc['期末','语文'])
```

```python
#显示构造


```

```python
#列表
indexs = pd.MultiIndex.from_arrays([['期中','期中','期中','期末','期末','期末'],['语文','数学','外语','语文','数学','外语']])
columns = ['tom','jack','rose']
data = np.random.randint(0,150,size=(6,3)) #6行三列
pd.DataFrame(data,index=indexs,columns=columns)
```

```python
#from_tuples，以元组构造
```

```python
#from_product最简单
indexs = pd.MultiIndex.from_product([['期中','期末'],['语文','数学','外语']])
columns = ['tom','jack','rose']
data = np.random.randint(0,150,size=(6,3)) #6行三列
df = pd.DataFrame(data,index=indexs,columns=columns)
```

```python
df
```

```python
#索引与切片


```

```python
#索引
print(df.loc['期中'])
print()
print(df.loc['期中','语文'])
print()
print(df.loc['期中'].loc['语文'])
print()
print(type(df.loc['期中']))
print()
print(df['tom']) #列不需要loc或iloc
```

```python
#切片
print(df.loc['期中':'期末'])
print()
print(df.iloc[1:2])
print()
print(df.iloc[0:4])
```

```python
#索引的堆
#stack把列索引变成行索引
data = np.random.randint(0,150,size=(2,8))
index = ['rose','jack']
columns = pd.MultiIndex.from_product([['期末','期中'],['一模','二模'],['语文','数学']])
df = pd.DataFrame(data,index=index,columns=columns)
```

```python
print('------------------df------------------')
print(df)
print('------------------df.stack()------------------')
print(df.stack())
print('------------------df.stack().stack()------------------')
print(df.stack().stack())
print('------------------df.stack().stack().stack()------------------')
print(df.stack().stack().stack())
```

```python
#stack+level就是把指定列索引变成行索引
print('------------------df------------------')
print(df)
print('------------------df.stack()------------------')
print(df.stack())
print('------------------df.stack(level=0)------------------')
print(df.stack(level=0))
print('------------------df.stack(level=-1)------------------')
print(df.stack(level=-1))
print('------------------df.stack(level=2)------------------')
print(df.stack(level=2))
```

```python
#unstack,把行索引变成列索引
print('------------------df------------------')
print(df)
print('------------------df.unstack()------------------')
print(df.unstack())
print('------------------df.stack(level=0)------------------')
print(df.unstack(level=0))
```

```python
#多索引重排


```

```python
#sort_index()
df = df.stack()
print(df)
print()
print(df.sort_index())
```

```python
#
print(df)
print()
print(df.swaplevel())
print()
print(df.swaplevel(axis=0))
```
