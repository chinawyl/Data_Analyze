```python
datas = datas.loc[:,['收看时间','回看时间','观看时间']]
```

```python
datas.head()
```

```python
#处理NaT


```

```python
datas['回看时间'] = np.where(pd.isnull(datas['回看时间']) == True, 0, datas['回看时间'])
```

```python
datas['收看时间'] = np.where(pd.isnull(datas['收看时间']) == True, 0, datas['收看时间'])
```

```python
datas['观看时间'] = np.where(pd.isnull(datas['观看时间']) == True, 0, datas['观看时间'])
```

```python
datas.head()
```

```python
#计算用户观看时间


```

```python
datas['用户总时间(天 时 分 秒)'] = datas['收看时间'] + datas['回看时间'] + datas['观看时间']
```

```python
#定义时间转换函数
def convert_timedelta(duration):
    days, seconds = duration.days, duration.seconds
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = (seconds % 60)
    return days, hours, minutes, seconds
```

```python
times = []
for i in datas.index:
    d,h,m,s = convert_timedelta(datas['用户总时间(天 时 分 秒)'][i])
    times.append(d*24*60 + h*60 + m + s/60)
```

```python
datas['用户总时间(分)'] = times
```

```python
datas['用户总时间(分)'] = datas['用户总时间(分)'].map(lambda x:('%.2f')%x)
```
