```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
'''
（1）请画出上证指数的2020年的收盘价，10日均线，30日均线。

（2）假如2005年初，你有10万元，通过买卖上证指数，到现在为止，你投资资金是多少？买点：MA10（10日均线）上穿MA30，卖点：MA10下穿MA30
'''
data = pd.read_csv('./股票作业.txt',sep = '\t',index_col = 0,parse_dates = True)
```

```python
data
```

```python
x = list(data.loc['2020-01-02':'2020-03-03',:].index)
```

```python
x1 = list(data.loc['2020-01-02':'2020-01-10',:].index)
x2 = list(data.loc['2020-01-11':'2020-01-23',:].index)
x3 = list(data.loc['2020-02-03':'2020-02-10',:].index)
x4 = list(data.loc['2020-02-11':'2020-02-28',:].index)
x5 = list(data.loc['2020-03-02':'2020-03-03',:].index)
```

```python
y1 = list(data.loc['2020-01-02':'2020-01-10',:]['close'])
y2 = list(data.loc['2020-01-11':'2020-01-23',:]['close'])
y3 = list(data.loc['2020-02-03':'2020-02-10',:]['close'])
y4 = list(data.loc['2020-02-11':'2020-02-28',:]['close'])
y5 = list(data.loc['2020-03-02':'2020-03-03',:]['close'])
```

```python
#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#绘制图片
plt.plot(x1,y1,label="1月10日均线",color="orange")
plt.plot(x2,y2,label="1月30日均线",color="cyan")
plt.plot(x3,y3,label="2月10日均线",color="blue")
plt.plot(x4,y4,label="2月28日均线",color="red")
plt.plot(x5,y5,label="3月10日均线",color="green")

#设置x轴的刻度
x_labels = [i for i in x]
plt.xticks(x,x_labels,fontproperties=my_font,rotation=90)

#绘制表格
plt.grid(alpha=1) #alpha是表格透明度

#添加描述信息
plt.xlabel("日期",fontproperties=my_font)
plt.ylabel("收盘价",fontproperties=my_font)
plt.title("上证指数的2020年的收盘价，10日均线，30日均线",fontproperties=my_font)

#添加图例
plt.legend(prop=my_font,loc="upper right") #loc默认为右上角

#保存图片
plt.savefig("./Practice_Age.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```

```python
data['MA10']= data['close'].rolling(window=10).mean()
data['MA30']= data['close'].rolling(window=30).mean()
```

```python
yes = data['MA10']<data['MA30']
data['MA10'] = data['MA10'].shift()
data['MA30'] = data['MA30'].shift()
today = data['MA10']<data['MA30']


buy = data[(yes==False) & (today==True)]
sell = data[(yes==True) & (today==False)]

sell = sell[1:]
buy = buy.reset_index()
sell = sell.reset_index()
new_data = pd.merge(buy, sell, left_index=True, right_index=True)


profit = (new_data['close_y']-new_data['close_x'])/new_data['close_x']
benjin = 100000
for i in profit:
    benjin  = benjin*(1+float(i))
print('投資資金為:',benjin)
```
